```python
#Question 7
restaurant=requests.get('https://www.dineout.co.in/pune-restaurants/welcome-back?p=1')
soup=BeautifulSoup(restaurant.content)
soup
```

```python
name=[]
for i in soup.find_all('div',class_="restnt-info cursor"):
    name.append(i.text)
name
```

```python
cuisine=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text)
cuisine
```

```python
ratings=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    ratings.append(i.text)
ratings
ratings.insert(8,'4.5')
ratings.insert(15,'4.7')
ratings.insert(16,'4.5')
ratings
```

```python
image=[]
for i in soup.find_all('img',class_="no-img"):
    image.append(i['data-src'])
image
```

```python
dineout=pd.DataFrame({'Restaurant Name':name,'cuisine1':cuisine,'Rating':ratings,'Image URL':image})
length=len(name),len(cuisine),len(ratings),len(image)
length
dineout
```

```python
dineout[['Price','Cuisine']]=dineout['cuisine1'].str.split(r'|',expand=True)
dineout
dineout.drop(['cuisine1'],axis=1)
```

```python
# Question 5
cnbc=requests.get('https://www.cnbc.com/world/?region=world')
soup=BeautifulSoup(cnbc.content)
soup
```
